```python
import random
def make_dic(deck1):
    deckdic1={'Joker':[0]}
    for card in deck1:
        if card !='Joker':
            c=card.split()
            key=c[1]
            if key in deckdic1:
                deckdic1[key].append(c[0])
            else:
                deckdic1[key]=[c[0]]
        else:
            deckdic1[card][0]+=1
    return deckdic1


def lying(lvl):
    shapes=['heart','spade','club','diamond']
    numbers=['Ace','2','3','4','5','6','7','8','9','10','Jack','Queen','King']
    joker='Joker'
    joker_n=3
    Deck=[]
    for number in numbers:
        for shape in shapes:
            Deck.append(shape+' '+number)
    for j in range(joker_n):
        Deck.append(joker)
    # print(Deck)
    n_shuffle=10
    for shuffle in range(n_shuffle):
        random.shuffle(Deck)
    amount_pp=15
    deck1=[]
    deck2=[]
    for card in range(amount_pp):
        deck1.append(Deck.pop())
        deck2.append(Deck.pop())
    # print(deck1)
    # print(deck2)

    deckdic1=make_dic(deck1)
    deckdic2=make_dic(deck2)
#     print(deckdic1)
    # print(deckdic2)
    most=max(len(shapes) for shapes in deckdic1.values())
#     print('most',most)
    best_n=[]

    for key,shape in deckdic1.items(): 
        if key != "Joker":
             if len(shape)==most:
                best_n.append(key)
#     print('available numbers: ', best_n)
    chosen_n=best_n[int(random.uniform(0,len(best_n)))]
#     print('chosen: ',chosen_n,deckdic1[chosen_n])
    print()

#     print('**************')
    print(drunkprompt('chosen: card ',max(0,lvl-1)),chosen_n )
    avail=deckdic1[chosen_n]
    cards_lay=0
    status='LOST'
    winner=False
    turns=5
    while cards_lay<turns and winner==False:
        print(drunkprompt('Number of turns to go before you lose: ',max(0,lvl-1)),turns-cards_lay)
        cards_lay+=1
        liechance=int(random.uniform(1,10))
        if liechance!=1 and len(avail)!=0:
            taken_shape=avail.pop(int(random.uniform(0,len(avail))))
        else:
            jokerchance=int(random.uniform(1,10))
            if jokerchance==5 and deckdic1['Joker'][0]!=0:
                taken_shape='Joker'
                deckdic1['Joker'][0]-=1
            else:
                taken_shape='Lie'
        ans=input('Lie? (yes/no) ')
        if ans=='yes' and (taken_shape!='Lie'):
#             print('YOU LOSE')
            status='LOST'

            winner=True
        elif ans=='yes' and (taken_shape=='Lie'):
#             print('YOU WIN')

            status='WON'
            winner=True
    print("*********************************")
    print('It was a', taken_shape)
    print('CONGRATS!') if status == "WON" else print('Better luck next time')
    return True if status == "WON" else False
```
