```python
class food(object):
    
    def __init__(self):
        self.cooktime=10
        self.burntime=10
        self.firetime=5
        self.cook_times=[self.cooktime, self.burntime, self.firetime]
        
        self.step_cook_state=0
        self.cook_states=["raw", "cooked", "burning","flaming","burned"]
        self.current_cook_state=self.cook_states[self.step_cook_state]
        self.cookmethod="Cooking"
        
        self.step_cut_state=0
        self.cut_states=["uncut", "cut"]
        self.current_cut_state=self.cut_states[self.step_cut_state]
        self.cutmethod="cutting"
        self.name="Unnamed food"
        

    def cut(self):
        if not self.step_cook_state:
            if self.step_cut_state<len(self.cut_states)-1:
                self.step_cut_state+=1
                self.current_cut_state=self.cut_states[self.step_cut_state]
                print(self.cutmethod)
            else:
                print("Already cut")
        else:
            print("Can't cut cooked food")
    
    def cook(self):
        if self.step_cut_state:
            if self.step_cook_state<len(self.cook_states)-2:
                print(self.cookmethod+"...")
                self.step_cook_state+=1
                self.current_cook_state= self.cook_states[self.step_cook_state]
                print(str(self.cook_times[self.step_cook_state-1])+" minutes")
                print(self.current_cook_state)
            elif self.step_cook_state==len(self.cook_states)-2:
                print("Flaming!")
            else:
                print("Burned. Throw away.")
        else:
            print("First needs to be cut!")
    
    def extinguish(self):
        if self.step_cook_state==len(self.cook_states)-2:
            self.step_cook_state+=1
            self.current_cook_state= self.cook_states[self.step_cook_state]
            print("Fire extinguished. Throw away.")
        else:
            print("Not flaming. So nothing to extinguish.")
            
    def throw_away(self):
        print("Throwing away "+str(self))
        self.step_cut_state=0
        self.current_cut_state=self.cut_states[self.step_cut_state]
        self.step_cook_state=0
        self.current_cook_state= self.cook_states[self.step_cook_state]
    
    def __str__(self):
        return self.current_cook_state + " "+self.current_cut_state +" " + self.name
    
class Brocoli(food):
    def __init__(self):
        super().__init__()
        self.name="brocoli"
        self.cooktime=15
        self.cook_times=[self.cooktime, self.burntime, self.firetime]
        self.cutmethod="Cutting in little stems"
        self.cookmethod="Cooking in water"
        

class Extinguisher():
    def __init__(self):
        self.name="Extinguisher"
    
    def extinguish(self):
        print("You cant extinguish an extinguisher you dummy.")
        
    def cut(self):
        print("Please dont cut an extinguisher. It will make a mess.")
        
    def cook(self):
        print("Please dont cook an extinguisher.")
        
class Player():
    def __init__(self):
        self.location_index=0
        self.locationlist=["entrance","fridge","cupboard","cuttingboard","cookingstand","Thrashcan"]
        self.location=self.locationlist[self.location_index]
        self.inventory={}
        
        self.ingredients={}
        self.ingredients["fridge"]=["brocoli"]
        self.ingredients["cupboard"]=["pasta"]
        self.ingredients["entrance"]=["Extinguisher"]
        
        
    def go_to(self):
        correct_index=False
        print(self.locationlist)
        move_index=int(input("Where do you want to go? (1-"+ str(len(self.locationlist))+") "))-1
        if move_index in range(len(self.locationlist)):
            correct_index=True
  
        while not correct_index:
            print("Number not in range")
            move_index=int(input("Where do you want to go? (1-"+str(len(self.locationlist))+") "))-1
            if move_index in range(len(self.locationlist)):
                correct_index=True

            
        self.location_index=move_index
        self.location=self.locationlist[self.location_index]
        print("Going to the "+self.location)
        

    def add(self, ingredient):
        if self.location in self.ingredients:
            if ingredient.name in self.ingredients[self.location]:
                print(ingredient.name + " added")
                if ingredient.name not in self.inventory:
                    self.inventory[ingredient.name]=ingredient
                else:
                    print(ingredient.name +" already added")
        else:
            print(ingredient.name + " is not here.")

    def check_ingredients(self):
        if self.location in self.ingredients:
            print(self.ingredients[self.location])
            return self.ingredients[self.location]
        else:
            print("Nothing to see here")
        
    def cut(self, ingredient):
        if self.location=="cuttingboard":
            if ingredient in self.inventory:
                self.inventory[ingredient].cut()
            else:
                print(ingredient.name + " not in inventory")
        else:
            print("You cant cut here. Go to cuttingboard")
            self.go_to()
    
    def cook(self, ingredient):
        if self.location=="cookingstand":
            if ingredient in self.inventory:
                self.inventory[ingredient].cook()
            else:
                print(ingredient + " not in inventory")
        else:
            print("You cant cook here. Go to cookstand")
            self.go_to()
    
    def extinguish(self, ingredient):
        if "Extinguisher" in self.inventory.keys():
            if ingredient in self.inventory:
                self.inventory[ingredient].extinguish()
            else:
                print(ingredient + " not in inventory")
        else:
            print("Extinguisher not in inventory! Go to the entrance!")
            self.go_to()
            
    def throw_away(self, ingredient):
        if self.location=="Thrashcan":
            print("Throwing away "+str(ingredient))
            del self.inventory[ingredient]
        else:
            print("You cant throw it away here. Go to thrashcan")
            self.go_to()
            
```

```python
import time
def overcooked_txt():  
    inst_dic={}
    inst_dic["brocoli"]=Brocoli()
    inst_dic["Extinguisher"]=Extinguisher()

    game_on=True
    print("Hello! Welcome to this new restaurant! You begin at the entrance. Let's cook some brocoli...")
    time.sleep(1)
    me=Player()
    while game_on:

        input_player1=input("Next action? (1)check (2)move (3)take (4)action (0)exit ")
        if input_player1=="0":
            game_on=False
            del me
            break
        elif input_player1=="1":
            input_player2=input("What do you want to check? (1)Location (2)Inventory ")
            if input_player2=="1":
                print("At this location: ")
                me.check_ingredients()
            elif input_player2=="2":
                print("In your inventory: ")
                print(str(list(me.inventory.keys())))
        elif input_player1=="2":
            me.go_to()
        elif input_player1=="3":
            ingredientlist=me.check_ingredients()
            if ingredientlist:
                input_player2=input("What do you want to take? (index 1->) ")
                me.add(inst_dic[ingredientlist[int(input_player2)-1]])
        elif input_player1=="4":
            input_player2=input("What do you want to do? (1) Cut (2) Cook (3) Extinguish (4)Throw away ")
            if input_player2=="1":
                input_player3=input("What do you want to cut? " +str(list(me.inventory.keys()))+" index(1->)")
                me.cut(list(me.inventory.keys())[int(input_player3)-1])
            elif input_player2=="2":
                input_player3=input("What do you want to cook? " +str(list(me.inventory.keys()))+" ")
                me.cook(list(me.inventory.keys())[int(input_player3)-1])
            elif input_player2=="3":
                input_player3=input("What do you want to extinguish? " +str(list(me.inventory.keys()))+" ")
                me.extinguish(list(me.inventory.keys())[int(input_player3)-1])
            elif input_player2=="4":
                input_player3=input("What do you want to throw away? " +str(list(me.inventory.keys()))+" ")
                me.throw_away(list(me.inventory.keys())[int(input_player3)-1])
            else:
                pass


```
